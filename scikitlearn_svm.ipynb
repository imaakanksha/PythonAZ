```python
#Use SVM on the iris dataset n we will create a model which can  classify the flowers on the basis of their features.
print("Starting new project on Scikit-Learn")
```

```python
#SVM Classifier
from sklearn import svm
from sklearn import datasets
iris=datasets.load_iris()
type(iris)
```

```python
iris.data
```

```python
iris.feature_names
```

```python
iris.target
```

```python
iris.target_names
```

```python
x=iris.data[:,2]
```

```python
y=iris.target
```

```python

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2, random_state=4)
```

```python
x_train_mod=x_train.reshape(-1,1)
```

```python
x_test_mod=x_test.reshape(-1,1)
```

```python
y_train_mod=y_train.reshape(-1,1)
```

```python
y_test_mod=y_test.reshape(-1,1)
```

```python
model = svm.SVC(kernel = 'linear')
```

```python
model.fit(x_train_mod,y_train_mod)
```

```python
y_prod_mod=model.predict(x_test_mod)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test_mod,y_prod_mod))
```
